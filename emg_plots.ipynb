Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Load continuous wrist flexion data from Delsys Galileo sensor

```python
df = pd.read_csv("galileo_data.csv", header=0)
df.columns = ["ch_1_time", "ch_1_voltage", 
              "ch_2_time", "ch_2_voltage", 
              "ch_3_time", "ch_3_voltage", 
              "ch_4_time", "ch_4_voltage"]
```

Plot

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(12, 12))
fig.suptitle("Wrist Flexion Measured With Delsys Galileo Sensor: Raw EMG Data")
fig.subplots_adjust(hspace=0.3)

ax1.plot(df["ch_1_time"], df["ch_1_voltage"])
ax1.set_title("Channel 1")
ax1.set_xlabel("Time (s)")
ax1.set_ylabel("Voltage (mV)")

ax2.plot(df["ch_2_time"], df["ch_2_voltage"])
ax2.set_title("Channel 2")
ax2.set_xlabel("Time (s)")
ax2.set_ylabel("Voltage (mV)")

ax3.plot(df["ch_3_time"], df["ch_3_voltage"])
ax3.set_title("Channel 3")
ax3.set_xlabel("Time (s)")
ax3.set_ylabel("Voltage (mV)")

ax4.plot(df["ch_4_time"], df["ch_4_voltage"])
ax4.set_title("Channel 4")
ax4.set_xlabel("Time (s)")
ax4.set_ylabel("Voltage (mV)")

plt.show()
```
